# HEART DISEASE PREDICTION
### INCLUDES PRINCIPAL COMPONENT ANALYSIS

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Samhith\OneDrive\Documents\studies\heart.csv")
df
```

```python
sns.scatterplot(df)
plt.show()
```

```python
sns.scatterplot(x="Age",y="Cholesterol",data=df)
plt.show()
```

```python
df["z-score"]=(df["Cholesterol"]-df["Cholesterol"].mean())/df["Cholesterol"].std()
df["z-score"]
```

```python
df.head()
```

## OUTLIER REMOVAL

```python
df1=df[(df["z-score"]>3) | (df["z-score"]<-3)]
df1
```

```python
df1=df[(df["z-score"]<3) & (df["z-score"]>-3)]
df1
```

```python
sns.scatterplot(x="RestingBP",y="Age",data=df1)
plt.show()
```

```python
sns.scatterplot(x="MaxHR",y="Age",data=df1)
plt.show()
```

```python
df1["z-score1"]=(df1["MaxHR"]-df1["MaxHR"].mean())/df1["MaxHR"].std()
df1["z-score1"]
```

```python
df1.head()
```

```python
df2=df1[(df1["z-score1"]>3) | (df1["z-score1"]<-3)]
df2
```

```python
df2=df1[(df1["z-score1"]<3) & (df1["z-score1"]>-3)]
df2
```

```python
df2.drop(["z-score","z-score1"],axis="columns",inplace=True)
df2
```

## DUMMY VARIABLESCREATION

```python
Sex_dummies=pd.get_dummies(df2["Sex"])
Sex_dummies
```

```python
CPT_dummies=pd.get_dummies(df2["ChestPainType"])
CPT_dummies
```

```python
ECG_dummies=pd.get_dummies(df2["RestingECG"])
ECG_dummies
```

```python
Exercise_dummies=pd.get_dummies(df2["ExerciseAngina"])
Exercise_dummies
```

```python
ST_dummies=pd.get_dummies(df2["ST_Slope"])
ST_dummies
```

```python
merged=pd.concat([df2,Sex_dummies,CPT_dummies,ECG_dummies,Exercise_dummies,ST_dummies],axis="columns")
merged
```

```python
merged.drop(["Sex","ChestPainType","RestingECG","ExerciseAngina","ST_Slope"],axis="columns",inplace=True)
merged
```

```python
merged.drop(["F","ASY","LVH","N","Down"],axis="columns",inplace=True)
merged
```

```python
X=merged.drop(["HeartDisease"],axis="columns")
X
```

```python
y=merged.HeartDisease
y
```

## SCALING THE DATA 

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scaled=scale.fit_transform(X)
scaled
```

## TRAINING THE DATASET

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
X_train.shape
```

```python
X_test.shape
```

## CHOOSING THE BEST MODEL

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
```

```python
model_params={
    "decisiontree":{
        "model":DecisionTreeClassifier(),
        "params":{
            "criterion":["gini","entropy"]
        }
    },
    "randomforest":{
        "model":RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10]
        }
    },
    "gaussianNB":{
        "model":GaussianNB(),
        "params":{}
    },
    "logisticRegression":{
        "model":LogisticRegression(),
        "params":{
            "C":[1,5,10]
        }
    },
    "svm":{
        "model":SVC(),
        "params":{
            "kernel":["rbf","linear"],
            "C":[1,10,20]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
scores=[]

for model_name,mp in model_params.items():
    clf=GridSearchCV(mp["model"],mp["params"],cv=5,return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({ 
    "model":model_name,
    "best param":clf.best_params_,
    "best score":clf.best_score_
})
```

```python
dataframe=pd.DataFrame(scores,columns=["model","best param","best score"])
dataframe
```
